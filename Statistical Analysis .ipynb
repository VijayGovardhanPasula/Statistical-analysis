```python
# Defining a function to get values from the raw data, using DateIndex from new DataFrame row
def populate_df(row):
    index = str(row['date'].year)
    value = raw_world.loc[index]
    return value
```

```python
# Select just the co2 emissions for the 'world', and only for the years 1960-2018 
raw_world = df_co2[df_co2['Country Name']=='World'].loc[:,'1960':'2018']
```

```python
# Traspose the result
raw_world = raw_world.T
raw_world.columns = ['value']
```

```python
# Create a new DataFrame with a daterange the same the range for the Temperature data (after resampling to years)
date_rng = pd.date_range(start='31/12/1960', end='31/12/2018', freq='y')
e = pd.DataFrame(date_rng, columns=['date'])
```

```python
# Populate the new DataFrame using raw data.
v = e.apply(lambda row: populate_df(row), axis=1)
e['Global CO2 Emissions per Capita'] = v
e.set_index('date', inplace=True)
```

```python
# Create figures and axes for plotting and plot co2 emissions data with green colour and line thickness
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(e, color='green', linewidth=2.4)

# Set axis labels and graph title
ax.set(xlabel='Time in years', ylabel='Emissions (Metric Tons per Capita)',
       title='Global CO2 Emission')
ax.grid()
```

```python
#reading the urban population data separately 
df_up = pd.read_excel('/Users/vijaygovardhanpasula/Downloads/Urban_total.xls',skiprows=3)
```

```python
#dropping all the unneccsary columns from the urban population 
df_up = df_up.drop(df_up.columns[4:34], axis = 1)
```

```python
#selecting only United Arab Emirates, Australia, Canada, USA, and UK
df_up_con = df_up[df_up['Country Code'].isin(["ARE", "AUS", "CAN", "USA", "GBR"])]
```

```python
#resetting the index
df_up_con = df_up_con.reset_index(drop = True)
```

```python
# Plotting the urban population for only selected countries
dfg = df_up_con.groupby(by='Country Name').sum()
dfg.sort_values(by=dfg.columns[-1], ascending=False, inplace=True)
dfplot = dfg.iloc[:5].T.plot()
plt.xlabel("Years")
plt.ylabel("% of Total Urban Population")
plt.title("Urban Population for 1990-2020")
plt.show()
```
